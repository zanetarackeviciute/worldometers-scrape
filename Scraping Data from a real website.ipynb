```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser', from_encoding='utf-8')
```

```python
print(soup)
```

```python
world_columns = soup.find_all('th')
```

```python
world_column_title = [columns.text.strip() for columns in world_columns]
print(world_column_title)
```

```python
import pandas as pd
df = pd.DataFrame(columns = world_column_title)
df
```

```python
world_rows = soup.find_all('tr')
```

```python
for row in world_rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
df
```
